```python
#getting rid of @,hahstags and html
import re 
#'' means char, does not create a space 
def regex_fix(x):
    x = re.sub("@(\w|\:|/)*", '', x)
    x = re.sub("http(\w|\:|/|\.)*", '', x)
    x = re.sub("#", '', x)
    x = re.sub("\\n",'', x)
    return x
    
raw_df["Text_cl"] = raw_df["Text"].apply(lambda x: regex_fix(x))

```

```python
# Function for converting emoticons into word
from emot.emo_unicode import EMOTICONS_EMO, EMOJI_UNICODE
def convert_emoticons(text):
    for emot in EMOTICONS_EMO:
        text = re.sub(re.escape(emot), ' ' + "_".join(EMOTICONS_EMO[emot].replace(",","").split()) + " ", text)
    return text

text = "Hello :-):-)"
convert_emoticons(text)
```

```python
def convert_emojis(text):
    for key in EMOJI_UNICODE:
        text = re.sub(re.escape(EMOJI_UNICODE[key]), " " + "_".join(key.replace(",","").replace(":","").split()) + " ", text)
    return text
convert_emojis("I won🥲🥰in")
```

```python
#emojie replacement 
def emo_rep(text):
    text = convert_emoticons(text)
    text = convert_emojis(text)
    #use two emoji libraries 
    return text
raw_df["Text-cl-1"] = raw_df.iloc[0:500,3].apply(lambda x: emo_rep(x))
raw_df.to_csv("emo_to_w.csv")
```

```python
#stopwords
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

def rm_stopw(text):
    text = text.split()
    return ' '.join([x for x in text if x not in stop_words])

raw_df["rm_sw"] = raw_df["Text-cl-1"].apply(lambda x: rm_stopw(x))
raw_df.to_csv("rm_sw")
```

```python
raw_df.head()
# #stemming 
from nltk.stem import PorterStemmer


def stem_sentences(text):
    tokens = text.split()
    stemmed_tokens = [PorterStemmer().stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)

raw_df['stemmed'] = raw_df["rm_sw"].apply(lambda x: stem_sentences(x))
raw_df.to_csv('stemmed.csv')
```

```python
from nltk.tokenize import word_tokenize

def tokenizer(text): 
    text = word_tokenize(text)
    return text
raw_df["tokens"] = raw_df["stemmed"].apply(lambda x: tokenizer(x))
raw_df.to_csv('tokens.csv')
```

```python
from nltk import ngrams 

def app_ngram(text, n):
    res = ngrams(text.split(), n)
    return [x for x in res]

raw_df["bi- grams"] = raw_df["stemmed"].apply(lambda x: app_ngram(x,2))
raw_df["tri- grams"] = raw_df["stemmed"].apply(lambda x: app_ngram(x,3))

raw_df.to_csv('ngrams.csv')
```
